## Correlação

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv('./2015-building-energy-benchmarking.csv')
data_set.head()
```

```python
data_set.drop(['DataYear'], axis=1,inplace = True)
data_set.corr(method='pearson')
```

- ignora as variáveis não numéricas(categóricas)
- id e date year não fazem sentido estarem no mapa de calor, pois não fazem sentido
- as variáveis mais "claras" poderiam ser retiradas, pois tem uma alta correlação

- valor de correlação varia entre -1 (correlação inversamente proporcional) e 1 (correlação proporcional)
- quando mais próximo de 0, menor correlação tem

```python
plt.figure(figsize=(20,20))
sns.heatmap(data_set.corr())
plt.show
```
