## Ajustando os Parâmetros da Regressão Logística

```python
import pandas as pd
pd.set_option('display.max_columns', 64)
pd.set_option('display.max_rows',64)
dataset_raw = pd.read_csv('./Data_train_reduced.csv')
```

```python
# dados faltantes muito acima de 20 % são retirados

dataset_raw.drop('q8.8', axis=1, inplace = True)
dataset_raw.drop('q8.9', axis=1, inplace = True)
dataset_raw.drop('q8.10', axis=1, inplace = True)
dataset_raw.drop('q8.20', axis=1, inplace = True)
dataset_raw.drop('q8.17', axis=1, inplace = True)
dataset_raw.drop('q8.18', axis=1, inplace = True)
dataset_raw.drop('q8.2', axis=1, inplace = True)
dataset_raw.drop('Respondent.ID', axis=1, inplace = True)
dataset_raw.drop('Product', axis=1, inplace = True)
dataset_raw.drop('q1_1.personal.opinion.of.this.Deodorant', axis=1, inplace=True)

# dados faltantes abaixo de 20%, são substituidos pela mediana
dataset_raw['q8.12'].fillna(dataset_raw['q8.12'].median(),inplace=True)
dataset_raw['q8.7'].fillna(dataset_raw['q8.7'].median(),inplace=True)
```

```python
y = dataset_raw['Instant.Liking']
x = dataset_raw.drop('Instant.Liking', axis=1)
```

### Variações de Parâmetros

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

valores_C = np.array([0.01, 0.1, 0.5, 1, 2, 3, 5, 10, 20, 50, 100])
regularizacao = ['l1', 'l2']
valores_grid = { 'C': valores_C, 'penalty': regularizacao }

modelo = LogisticRegression()

grid_regressao_logistica = GridSearchCV( estimator = modelo, param_grid = valores_grid, cv = 10)
grid_regressao_logistica.fit(x,y)

print("Melhor acurácia: ", grid_regressao_logistica.best_score_)
print("Parâmetro C: ", grid_regressao_logistica.best_estimator_.C)
print("Regularização: ", grid_regressao_logistica.best_estimator_.penalty)
```
