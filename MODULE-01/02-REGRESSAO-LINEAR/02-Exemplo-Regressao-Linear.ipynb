# Exemplo Regressão Linear

## Importar libs

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.model_selection import train_test_split
```

## Create Data
- make_regression, cria dados de forma aleatória (tipo regressão) com:
    - n_samples = quantidade de dados
    - n_features = x, y(2 dimensões)
    - noise = ruídos (adiciona mais dados menos correlacionados - quanto mais ruído, menos aparente é uma reta)

```python
x,y = make_regression(n_samples = 200, n_features = 1, noise = 30)
```

## View Data

```python
plt.scatter(x, y)
plt.show()
```

## Load LinearRegression
https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
modelo_teste = LinearRegression()
```

```python
modelo_teste.fit(x, y)
```

```python
modelo_teste.intercept_
```

```python
modelo_teste.coef_
```

## Plot modelo_teste


```python
# plot meus dados
plt.scatter(x, y)
# Valores de intervalos para x no gráfico acima
xreg = np.arange(-3 , 3 , 1)
```

```python
# Gráfico da regressão
# a regressão linear tem o seguinte formato : 
# y = ax + b
# a = modelo_teste.coef_
# b = modelo_teste.intercept_
plt.scatter(x, y)
plt.plot(xreg,68.58 * xreg-1.103, color= 'red')
plt.show()
# plot minha reta e meus dados
```

## Split Data

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.30)
```

```python
modelo_teste.fit(x_treino, y_treino)
```

```python
# retornando o coeficiente de determinação R2 em relação aos dados de teste
# modelo_teste é uma reta
# score é a função que retorna R2
resultado_r2 = modelo_teste.score(x_teste, y_teste)
```

```python
print(resultado_r2)
```
